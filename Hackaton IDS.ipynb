```python
import pandas as pd

```

```python
# df = pd.read_csv('vlucht data schiphol.csv')

df = pd.read_csv('flights_today_master.csv')

df.info()
```

```python
df['path'] = df['Latitude'].astype('string') + ", " + df['Longitude'].astype('string')


df.head()
```

```python
# Convert 'date' column to datetime
df['ScrapeTime'] = pd.to_datetime(df['ScrapeTime'])  

# Extract only the date part (YYYY-MM-DD)
df['date_str'] = df['ScrapeTime'].dt.strftime('%Y-%m-%d')

# Concatenate 'date_str' with 'time' and convert to datetime
df['datetime'] = pd.to_datetime(df['date_str'] + ' ' + df['Time'], format='%Y-%m-%d %a %I:%M:%S %p')

# Convert to timestamp (milliseconds)
df['timestamp'] = df['datetime'].apply(lambda x: int(x.timestamp() * 1000))

df['path'] = df['path'].tolist()

df.info()
```

```python
# Create a 'coordinates' column as a list [Latitude, Longitude]
df["coordinates"] = df.apply(lambda row: [row["Latitude"], row["Longitude"]], axis=1)

# Normalize timestamps by subtracting 1554772579000
df["timestamps"] = df["timestamp"] - 1554772579000

df.head()
```

```python
mile_in_km = 1.609344
df['Speed_kph'] = (df['Speed_mph'] * mile_in_km).round(2)
df = df.drop(columns=['Speed_mph'])
df.head()
```
